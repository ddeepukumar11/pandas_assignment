```python
d = dict()
for index, row in match_df.iterrows():
    venue = row['venue']
    if venue not in d:
        d[venue]=0
    if row['result'] == 'normal':
        if row['winner'] == row['team1']:
            d[venue]+=1
        elif row['winner'] == row['team2']:
            d[venue]-=1
```

```python
d
```

```python
for i in d:
    if d[i] > 0:
        d[i] = 'bat'
    elif d[i] < 0:
        d[i] = 'field'
    else:
        d[i] = 'bat/field'
```

```python
#best decision for each groound
```

```python
runs=dict()
for index, row in delivery_df.iterrows():
    batsman_runs = row['batsman_runs']
    bowler = row['bowler']
    if batsman_runs not in runs:
        runs[batsman_runs]=dict()
    if bowler not in runs[batsman_runs]:
        runs[batsman_runs][bowler]=0
    runs[batsman_runs][bowler]+=1
```

```python
runs
```
