```python
new_df = pd.read_csv('testset.csv')
```

```python
new_df.head()
```

```python
new_df = new_df.loc[:,'datetime_utc':' _conds']
```

```python
new_df.head()
```

```python
new_df['datetime_utc'] = pd.to_datetime(new_df['datetime_utc'])
```

```python
new_df = new_df.set_index('datetime_utc')
```

```python
new_df
```

```python
conditions = dict()
```

```python
for index,row in new_df.iterrows():
    yr = index.year
    month = index.month
    cond = row[' _conds']
    if yr == 2016 and month == 2:
        if cond not in conditions:
            conditions[cond]=0
        conditions[cond]+=1
```

```python
plt.pie(list(conditions.values()),labels=conditions.keys())
```
